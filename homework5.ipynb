```python
def sum_numbers_in_string(line):
    total = 0
    current_num = ''
    for char in line:
        if char.isdigit():
            current_num += char
        elif current_num:
            total += int(current_num)
            current_num = ''
    if current_num:
        total += int(current_num)
    return total

def replace_spaces_with_commas(line):
    return line.replace(' ', ', ')

def count_words_same_start_end(line):
    count = 0
    words = line.rstrip('.').split(',')
    for word in words:
        word = word.strip()
        if word and word[0].lower() == word[-1].lower():
            count += 1
    return count

def process_file(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        for i, line in enumerate(file, 1):
            line = line.strip()
            if not line:
                continue

            print(f"Строка {i}:")
            print(f"1. Сумма чисел: {sum_numbers_in_string(line)}")
            print(f"2. Строка с запятыми: {replace_spaces_with_commas(line)}")
            print(f"3. Количество слов: {count_words_same_start_end(line)}\n")

if __name__ == "__main__":
    process_file('input.txt')
```

A1,    AN,    A2,    AN–1,    A3,    … .
3.	Дан файл вещественных чисел. Создать два новых файла, первый из которых содержит элементы исходного файла с нечетными номерами (1, 3, …), а второй — с четными (2, 4, …).


```python
def remove_before_first_space(input_file, output_file):
    with open(input_file, 'r') as f:
        content = f.read()

    space_pos = content.find(' ')
    result = content[space_pos+1:] if space_pos != -1 else content

    with open(output_file, 'w') as f:
        f.write(result)

def rearrange_numbers(input_file, output_file):
    with open(input_file, 'r') as f:
        numbers = list(map(int, f.read().split()))

    if not numbers:
        return

    N = len(numbers)
    new_seq = []
    for i in range(N):
        if i % 2 == 0:
            new_seq.append(numbers[N - 1 - i//2])
        else:
            new_seq.append(numbers[i//2])

    with open(output_file, 'w') as f:
        f.write(' '.join(map(str, new_seq)))

def split_odd_even_positions(input_file, odd_file, even_file):
    with open(input_file, 'r') as f:
        numbers = list(map(float, f.read().split()))

    with open(odd_file, 'w') as f:
        f.write(' '.join(map(str, numbers[::2])))

    with open(even_file, 'w') as f:
        f.write(' '.join(map(str, numbers[1::2])))

def main():
    remove_before_first_space('input1.txt', 'output1.txt')
    rearrange_numbers('input2.txt', 'output2.txt')
    split_odd_even_positions('input3.txt', 'odd.txt', 'even.txt')

if __name__ == "__main__":
    main()
```
