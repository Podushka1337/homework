1.	Дан одномерный массив В, состоящий из 2n элементов. Переставить его элементы по следующему правилу: второй и минимальный.
2.	Дан массив размера N. Найти максимальный из его элементов, не являющихся ни локальным минимумом, ни локальным максимумом. Если таких элементов в массиве нет, то вывести 0.
3.	Дан целочисленный массив размера N. Удалить из массива все нечетные числа и вывести размер полученного массива и его содержимое.
4.	Дан целочисленный двумерный массив, размерности n х m. Вывести на экран те элементы, у которых остаток от деления на m равен k.


```python
#1
def rearrange_array(arr):
    n = len(arr) // 2
    result = []
    for i in range(n):
        result.append(arr[i+1])
        result.append(min(arr[i], arr[i+1]))
    return result

B = list(map(int, input("Введите элементы массива через пробел (должно быть четное количество): ").split()))
print("Преобразованный массив:", rearrange_array(B))
```

```python
#2
def find_special_max(arr):
    if len(arr) < 3:
        return 0

    extremums = set()
    for i in range(1, len(arr)-1):
        if arr[i] > arr[i-1] and arr[i] > arr[i+1]:
            extremums.add(i)
        elif arr[i] < arr[i-1] and arr[i] < arr[i+1]:
            extremums.add(i)

    candidates = [arr[i] for i in range(len(arr)) if i not in extremums]

    return max(candidates) if candidates else 0

N = list(map(int, input("Введите элементы массива через пробел: ").split()))
print("Максимальный неэкстремальный элемент:", find_special_max(N))
```

```python
#3
def remove_odd_numbers(arr):
    even_numbers = [x for x in arr if x % 2 == 0]
    return len(even_numbers), even_numbers

N_array = list(map(int, input("Введите целые числа через пробел: ").split()))
size, new_array = remove_odd_numbers(N_array)
print("Новый размер:", size)
print("Только четные элементы:", new_array)
```

```python
#4
def find_elements_with_remainder(matrix, m, k):
    result = []
    for row in matrix:
        for element in row:
            if element % m == k:
                result.append(element)
    return result

n = int(input("Введите количество строк: "))
m = int(input("Введите количество столбцов: "))
matrix = []
for i in range(n):
    row = list(map(int, input(f"Введите {m} элементов {i+1}-й строки через пробел: ").split()))
    matrix.append(row)
k = int(input("Введите искомый остаток k: "))

result = find_elements_with_remainder(matrix, m, k)
print("Элементы с заданным остатком:", result)
```

Пересечение словарей: Напишите функцию, которая принимает два словаря и возвращает новый словарь, содержащий только те пары ключ-значение, которые присутствуют в обоих словарях.

```python
def input_dict():
    print("\nСоздать словарь:")
    result = {}
    while True:
        key = input("Ключ (или 'стоп'): ")
        if key == 'стоп':
            break
        value = input("Значение: ")
        result[key] = value
    return result

print("Первый словарь:")
dict1 = input_dict()

print("\nВторой словарь:")
dict2 = input_dict()

common = {}
for key in dict1:
    if key in dict2 and dict1[key] == dict2[key]:
        common[key] = dict1[key]

print("\nОбщие элементы:")
if common:
    for key, value in common.items():
        print(f"{key}: {value}")
else:
    print("Нет общих элементов")
```

```python
def word_frequency():
    text = input("Введите текст для анализа: ")
    words = text.split()
    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1
    print("Частота слов:", frequency)

word_frequency()
```

```python
def invert_dictionary():
    d = eval(input("Введите словарь в формате {'ключ1': значение1, 'ключ2': значение2}: "))
    inverted = {}
    for key, value in d.items():
        if value in inverted:
            if isinstance(inverted[value], list):
                inverted[value].append(key)
            else:
                inverted[value] = [inverted[value], key]
        else:
            inverted[value] = key
    print("Инвертированный словарь:", inverted)

invert_dictionary()
```

```python
def merge_dictionaries():
    d1 = eval(input("Введите первый словарь: "))
    d2 = eval(input("Введите второй словарь: "))
    merged = d1.copy()
    for key, value in d2.items():
        merged[key] = merged.get(key, 0) + value
    print("Объединенный словарь:", merged)

merge_dictionaries()
```

```python
def group_keys_by_length():
    d = eval(input("Введите словарь: "))
    result = {}
    for key in d.keys():
        length = len(key)
        if length not in result:
            result[length] = []
        result[length].append(key)
    print("Группировка ключей по длине:", result)

group_keys_by_length()
```

```python
# 1. Создание множества из списка
def list_to_set(input_list):
    return set(input_list)

# 2. Удаление дубликатов из списка с помощью множества
def remove_duplicates(input_list):
    return list(set(input_list))

# 3. Пересечение множеств
def set_intersection(set1, set2):
    return set1 & set2  # или set1.intersection(set2)

# 4. Объединение множеств
def set_union(set1, set2):
    return set1 | set2  # или set1.union(set2)

# 5. Разность множеств
def set_difference(set1, set2):
    return set1 - set2  # или set1.difference(set2)

# 6. Симметрическая разность
def symmetric_difference(set1, set2):
    return set1 ^ set2  # или set1.symmetric_difference(set2)

# 7. Проверка подмножества
def is_subset(subset, superset):
    return subset.issubset(superset)

# 8. Проверка надмножества
def is_superset(superset, subset):
    return superset.issuperset(subset)

# 9. Добавление элементов из списка в множество
def add_elements_to_set(input_set, elements_list):
    input_set.update(elements_list)
    return input_set

# 10. Удаление из множества элементов, присутствующих в списке
def remove_elements_from_set(input_set, elements_list):
    input_set.difference_update(elements_list)
    return input_set

# Примеры использования функций
if __name__ == "__main__":
    # Пример для 1-й функции
    my_list = [1, 2, 2, 3, 4, 4, 5]
    print("1. Множество из списка:", list_to_set(my_list))

    # Пример для 2-й функции
    print("2. Список без дубликатов:", remove_duplicates(my_list))

    # Примеры для остальных функций
    set_a = {1, 2, 3, 4, 5}
    set_b = {4, 5, 6, 7, 8}
    print("3. Пересечение множеств:", set_intersection(set_a, set_b))
    print("4. Объединение множеств:", set_union(set_a, set_b))
    print("5. Разность множеств:", set_difference(set_a, set_b))
    print("6. Симметрическая разность:", symmetric_difference(set_a, set_b))
    print("7. Является ли {1, 2} подмножеством:", is_subset({1, 2}, set_a))
    print("8. Является ли set_a надмножеством {1, 2}:", is_superset(set_a, {1, 2}))

    my_set = {1, 2, 3}
    add_elements_to_set(my_set, [3, 4, 5])
    print("9. После добавления элементов:", my_set)

    remove_elements_from_set(my_set, [1, 4])
    print("10. После удаления элементов:", my_set)
```

```python
# 1. Создание кортежа из списка
def list_to_tuple(input_list):
    return tuple(input_list)

# 2. Объединение кортежей
def tuple_concat(tuple1, tuple2):
    return tuple1 + tuple2

# 3. Проверка на наличие элемента
def element_in_tuple(input_tuple, element):
    return element in input_tuple

# 4. Индекс первого вхождения элемента
def first_index_of(input_tuple, element):
    return input_tuple.index(element)

# 5. Количество вхождений элемента
def count_element(input_tuple, element):
    return input_tuple.count(element)

# 6. Максимальный элемент
def tuple_max(input_tuple):
    return max(input_tuple)

# 7. Минимальный элемент
def tuple_min(input_tuple):
    return min(input_tuple)

# 8. Сортировка кортежа (возвращает список)
def sorted_tuple(input_tuple):
    return sorted(input_tuple)

# 9. Разбиение кортежа на две части
def split_tuple(input_tuple):
    half = len(input_tuple) // 2
    return (input_tuple[:half], input_tuple[half:])

# 10. Обратный порядок элементов
def reverse_tuple(input_tuple):
    return input_tuple[::-1]

# Примеры использования функций
if __name__ == "__main__":
    # Пример для 1-й функции
    my_list = [1, 2, 3, 4, 5]
    my_tuple = list_to_tuple(my_list)
    print("1. Кортеж из списка:", my_tuple)

    # Пример для 2-й функции
    tuple1 = (1, 2, 3)
    tuple2 = (4, 5, 6)
    print("2. Объединение кортежей:", tuple_concat(tuple1, tuple2))

    # Пример для 3-й функции
    print("3. Есть ли 3 в кортеже:", element_in_tuple(my_tuple, 3))
    print("   Есть ли 7 в кортеже:", element_in_tuple(my_tuple, 7))

    # Пример для 4-й функции
    print("4. Индекс первого вхождения 4:", first_index_of(my_tuple, 4))

    # Пример для 5-й функции
    tuple_with_duplicates = (1, 2, 2, 3, 2, 4)
    print("5. Количество вхождений 2:", count_element(tuple_with_duplicates, 2))

    # Пример для 6-й функции
    print("6. Максимальный элемент:", tuple_max(my_tuple))

    # Пример для 7-й функции
    print("7. Минимальный элемент:", tuple_min(my_tuple))

    # Пример для 8-й функции
    unsorted_tuple = (3, 1, 4, 2, 5)
    print("8. Отсортированный список:", sorted_tuple(unsorted_tuple))

    # Пример для 9-й функции
    long_tuple = (1, 2, 3, 4, 5, 6, 7, 8)
    part1, part2 = split_tuple(long_tuple)
    print("9. Разбиение кортежа:", part1, part2)

    # Пример для 10-й функции
    print("10. Обратный порядок:", reverse_tuple(my_tuple))
```
