5) ﻿﻿﻿Поле шахматной доски определяется парой натуральных чисел, каждое из которых не превосходит восьми: первое число — номер вертикали (при счете слева направо), второе — номер горизонтали (при счете снизу вверх). Даны натуральные числа а, b, с, d, е, f, каждое из которых не превосходит восьми. На поле (а, b) расположена белая фигура, на поле (с, d) — черная. Определить, может ли белая фигура пойти на поле (е, f), не попав при этом под удар черной фигуры.
Рассмотреть следующие варианты сочетаний белой и черной фигур: а) слон и ладья; б) король и слон, в) король и ферзь;

```python
def king_check(a, b, e, f):
    return abs(a - e) <= 1 and abs(b - f) <= 1

def queen(c, d, e, f):
    return c == e or d == f or abs(c - e) == abs(d - f)

a, b = map(int, input("координаты белого короля (a b): ").split())
c, d = map(int, input("координаты черного ферзя (c d): ").split())
e, f = map(int, input("координаты целевого поля (e f): ").split())

if king_check(a, b, e, f):
    if not queen(c, d, e, f):
        print("Белый король может пойти на поле (e, f), не попадая под удар ферзя.")
    else:
        print("Белый король может пойти на поле (e, f), но попадет под удар ферзя.")
else:
    print("Белый король не может пойти на поле (e, f).")
```

```python
N = int(input("введите n "))

if N <= 1:
    print("oшибка, n должно быть больше 1")
else:
    K = 0
    prev_num = int(input())

    for i in range(2, N + 1):
        current_num = int(input())

        if prev_num < current_num:
            print(f"Элемент {prev_num} меньше своего правого соседа.")
            K += 1

        prev_num = current_num

    print(f"Количество элементов, которые меньше своего правого соседа: {K}")
```

```python
number = int(input("Введите число: "))
count = 0

while number > 0:
    sum_digits = sum(int(digit) for digit in str(abs(number)))
    number -= sum_digits
    count += 1
print(f"Количество действий до получения нуля: {count}")
```

```python
N = int(input("введите размер: "))

replace_count = 0

for i in range(N):
    element = int(input())

    if element < 0:
        element = -element
        replace_count += 1
    print(f"текущий элемент после обработки: {element}")

# Вывод количества произведенных замен
print(f"количество произведенных замен: {replace_count}")
```

```python
import math

def sum_of_factorials(n):
    total_sum = 0
    for i in range(1, n + 1, 2):
        total_sum += math.factorial(i)
    return total_sum


n = int(input("введите нечетное число n (n <= 20): "))

if n % 2 == 0 or n > 20:
    print("ошибка: n должно быть нечетным и не превышать 20")
else:
    result = sum_of_factorials(n)
    print(f"Сумма факториалов нечетных чисел от 1 до {n}: {result}")
```
